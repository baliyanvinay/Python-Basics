```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
# because we have to remember the index or row of each country if we want to filter any row
df_can.set_index('Country', inplace=True)
# tip: The opposite of set is reset. So to reset the index, we can use df_can.reset_index()
```

```python
df_can.columns = list(map(str, df_can.columns)) # to convert year from int to string to avoid confusion
```

```python
# useful for plotting later on
years = list(map(str, range(1980, 2014)))
```

# Comparing the data for India and China

```python
# Getting data for India and China
India=df_can.loc['India',years]
China=df_can.loc['China',years]
print(India.head())
print(China.head())
```

```python
India.plot(kind='line')
China.plot
plt.show()
```

```python
# What if both countries are need to plotted in same graph?
India_China=df_can.loc[['India','China'],years]
```

```python
India_China.head()
```

```python
India_China.plot(kind='line')
plt.show()
```

```python
# the above chart is displayed this way because the column has years and rows has countries; this has to be transposed
India_China=India_China.transpose()
India_China.head()
```

```python
India_China.plot(kind='line')
plt.title('India-China Immigration Comparsion')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.show()
```

# Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
# sorted the data frame by total column in descending order
```

```python
# get the top 5 entries
df_top5 = df_can.head(5)
```

```python
df_top5.head()
```

```python
df_top5 = df_top5[years].transpose() 
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(15, 5)) # pass a tuple (x, y) size

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
